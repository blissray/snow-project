## 데이터 불러오기

```python
df = pd.read_csv('../../../data/데이터정리/전체_권역별_데이터.csv')
df['날짜'] = df['날짜'].map(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').date())
```

```python
df.head()
```

```python
df_도심권 = df[df.권역 == '도심권'].reset_index(drop=True)
df_동남권 = df[df.권역 == '동남권'].reset_index(drop=True)
df_동북권 = df[df.권역 == '동북권'].reset_index(drop=True)
df_서남권 = df[df.권역 == '서남권'].reset_index(drop=True)
df_서북권 = df[df.권역 == '서북권'].reset_index(drop=True)
```

## 도심권
- 눈이 온 경우에만 분석에 이용
- 시간에 따라 달라지는 기온, 적설량, 교통사고수, 차량통행속도, 교통량 분석

```python
df_도심권 = df_도심권[df_도심권['강설량_체크']==True]
df_도심권 = df_도심권.groupby(['시'])['기온(°C)', '계산된적설량(cm)', '교통사고수','차량통행속도', '교통량'].mean().reset_index()
```

```python
df_도심권
```

```python
ax = sns.lineplot(x="시", y="기온(°C)", data=df_도심권, color='r', ci='sd',label='기온(°C)')
ax2 = ax.twinx()
sns.lineplot(x="시", y="계산된적설량(cm)", data=df_도심권, color='b', ci='sd',label='적설량(cm)')
# ax.figure.legend()

ax.grid(True)
ax2.grid(True)

plt.title("도심권 시간에 따른 기온 및 적설량 변화", fontsize=20,color='black')

ax.set_xlabel('시간', fontsize=20, color='black')

ax.set_ylabel("기온(°C)",fontsize=20,color='red')
ax2.set_ylabel("계산된적설량(cm)",fontsize=20,color='blue')

plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['lines.linewidth'] = 1

ax.legend(loc='lower left', bbox_to_anchor=(0.675,0.9))
ax2.legend()

fig = plt.gcf()
plt.show()
fig.savefig('../result/시간_강설량_변화_그래프/도심권_기온_적설량_graph.png')
```

```python
ax = sns.lineplot(x="시", y="차량통행속도", data=df_도심권, color='r', ci='sd',label='차량통행속도')
ax2 = ax.twinx()
sns.lineplot(x="시", y="계산된적설량(cm)", data=df_도심권, color='b', ci='sd',label='적설량(cm)')
# ax.figure.legend()

ax.grid(True)
ax2.grid(True)

plt.title("도심권 시간에 따른 차량통행속도 및 적설량 변화", fontsize=20,color='black')

ax.set_xlabel('시간', fontsize=20, color='black')

ax.set_ylabel("차량통행속도",fontsize=20,color='red')
ax2.set_ylabel("계산된적설량(cm)",fontsize=20,color='blue')

plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['lines.linewidth'] = 1

ax.legend(loc='lower left', bbox_to_anchor=(0.675,0.9))
ax2.legend()

fig = plt.gcf()
plt.show()
fig.savefig('../result/시간_강설량_변화_그래프/도심권_차량통행속도_적설량_graph.png')
```

```python
ax = sns.lineplot(x="시", y="교통량", data=df_도심권, color='r', ci='sd',label='교통량')
ax2 = ax.twinx()
sns.lineplot(x="시", y="계산된적설량(cm)", data=df_도심권, color='b', ci='sd',label='적설량(cm)')
# ax.figure.legend()

ax.grid(True)
ax2.grid(True)

plt.title("도심권 시간에 따른 교통량 및 적설량 변화", fontsize=20,color='black')

ax.set_xlabel('시간', fontsize=20, color='black')

ax.set_ylabel("교통량",fontsize=20,color='red')
ax2.set_ylabel("계산된적설량(cm)",fontsize=20,color='blue')

plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['lines.linewidth'] = 1

ax.legend(loc='lower left', bbox_to_anchor=(0.675,0.9))
ax2.legend()

fig = plt.gcf()
plt.show()
fig.savefig('../result/시간_강설량_변화_그래프/도심권_교통량_적설량_graph.png')
```
