```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../../../data/데이터정리/전체데이터.csv')
df['날짜'] = df['날짜'].map(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').date())
df.head()
```

```python
df_all = df.groupby(['날짜','년','월','일','시']).mean().reset_index()
```

```python
df_all.describe()
```

```python
df_no_snow = df_all[df_all['계산된적설량(cm)']<=0]
df_no_snow.head()
```

```python
df_no_snow.shape
```

```python
df_no_snow.mean()
```

```python
df_little_snow = df_all[(df_all['계산된적설량(cm)']>0) & (df_all['계산된적설량(cm)']<5)]
df_little_snow.head()
```

```python
df_little_snow.shape
```

```python
df_little_snow.mean()
```

```python
df_much_snow = df_all[df_all['계산된적설량(cm)']>=5]
df_much_snow.head()
```

```python
df_much_snow.shape
```

```python
df_much_snow.mean()
```

```python
print("   눈이 안 내렸을 때 차량통행속도 평균 :", df_no_snow.mean()['차량통행속도'])
print("눈이 0~5cm 내렸을 때 차량통행속도 평균 :", df_little_snow.mean()['차량통행속도'])
print("눈이 5cm이상 내렸을 때차량통행속도 평균 :", df_much_snow.mean()['차량통행속도'])
```
