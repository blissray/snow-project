```python
import pandas as pd
import datetime
```

```python
# 그래프를 노트북 안에 그리기 위해 설정
%matplotlib inline

# 필요한 패키지와 라이브러리를 가져옴
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
print ('설정 파일 위치: ', mpl.matplotlib_fname())
path = '/home/jongsky/anaconda3/envs/python_3/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/NanumGothicEco.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
print(font_name)
plt.rc('font', family=font_name)
```

```python
df = pd.read_csv('../../../data/데이터정리/전체데이터.csv')
df['날짜'] = df['날짜'].map(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').date())
df.head()
```

```python
df['계산된적설량(cm)'].unique()
```

```python
df.groupby(['계산된적설량(cm)']).mean()
```

```python
df.groupby(['계산된적설량(cm)']).mean().describe()
```

```python
graph_df = pd.DataFrame(df.groupby(['계산된적설량(cm)']).mean()['기온(°C)'])
graph_df
```

```python
ax = sns.lineplot(x=graph_df.index, y="기온(°C)", data=graph_df, color='r', ci='sd',label='기온(°C)')
# ax.figure.legend()

ax.grid(True)

plt.title("온도와 강설량 상관도 그래프", fontsize=20,color='black')

ax.set_xlabel('계산된적설량(cm)', fontsize=20, color='black')

ax.set_ylabel("기온(°C)",fontsize=20,color='red')

plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams['lines.linewidth'] = 1

ax.legend(loc='lower left', bbox_to_anchor=(0.675,0.9))

fig = plt.gcf()
plt.show()
fig.savefig('../result/온도_강설량_상관도그래프/snow_tem_graph_change_std.png')
```
