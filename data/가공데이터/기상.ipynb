## 기본 DataFrame 불러오기

```python
basic_df = pd.read_csv('basic_dataframe.csv', encoding='cp949')

basic_df.날짜 = basic_df.날짜.map(lambda x : datetime.datetime.strptime(x, "%Y-%m-%d"))
basic_df['년'] = basic_df.날짜.map(lambda x : x.year)
basic_df['월'] = basic_df.날짜.map(lambda x : x.month)
basic_df['일'] = basic_df.날짜.map(lambda x : x.day)
basic_df['날짜'] = basic_df.날짜.map(lambda x : datetime.date(x.year, x.month, x.day))
```

```python
서울시_일별_기상관측_2017 = pd.read_csv('../원본데이터/기상/서울시_일별_기상관측_2017.csv')
서울시_일별_기상관측_2018 = pd.read_csv('../원본데이터/기상/서울시_일별_기상관측_2018.csv')
서울시_일별_기상관측 = pd.merge(서울시_일별_기상관측_2017, 서울시_일별_기상관측_2018, how='outer')
del 서울시_일별_기상관측['Unnamed: 0']

서울시_일별_기상관측.날짜 = 서울시_일별_기상관측.날짜.map(lambda x : x[:-4])
서울시_일별_기상관측.날짜 = 서울시_일별_기상관측.날짜.map(lambda x : datetime.datetime.strptime(x, "%Y년 %m월%d일"))
서울시_일별_기상관측['년'] = 서울시_일별_기상관측.날짜.map(lambda x : x.year)
서울시_일별_기상관측['월'] = 서울시_일별_기상관측.날짜.map(lambda x : x.month)
서울시_일별_기상관측['일'] = 서울시_일별_기상관측.날짜.map(lambda x : x.day)
서울시_일별_기상관측['날짜'] = 서울시_일별_기상관측.날짜.map(lambda x : datetime.date(x.year, x.month, x.day))

서울시_일별_기상관측 = 서울시_일별_기상관측[['날짜','년','월','일','날씨']]
```

## 구별_일별_기상관측 데이터 불러오기 (기온, 강수량 데이터)
#### 현충원 -> 동작구, 북악산 -> 종로구

```python
구별_코드번호 = {400 : '강남', 402 : '강동', 424 : '강북', 404 : '강서', 509 : '관악', 413 : '광진', 423 : '구로'
          , 417 : '금천', 410 : '기상청', 425 : '남현', 407 : '노원', 406 : '도봉', 408 : '동대문', 411 : '마포'
          , 422 : '북악산', 420 : '북한산', 412 : '서대문', 401 : '서초', 421 : '성동', 414 : '성북', 403 : '송파'
          , 405 : '양천', 510 : '영등포', 415 : '용산', 416 : '은평', 419 : '중구', 409 : '중랑', 418 : '한강', 889 : '현충원'}
```

```python
구별_일별_기상관측_2017 = pd.read_csv('../원본데이터/기상/구별_시간별_기상관측_2017.csv', encoding='cp949')
구별_일별_기상관측_2018 = pd.read_csv('../원본데이터/기상/구별_시간별_기상관측_2018.csv', encoding='cp949')

구별_일별_기상관측 = pd.merge(구별_일별_기상관측_2017, 구별_일별_기상관측_2018, how='outer')

구별_일별_기상관측.일시 = 구별_일별_기상관측.일시.map(lambda x : datetime.datetime.strptime(x, "%Y-%m-%d %H:00"))
구별_일별_기상관측['년'] = 구별_일별_기상관측.일시.map(lambda x : x.year)
구별_일별_기상관측['월'] = 구별_일별_기상관측.일시.map(lambda x : x.month)
구별_일별_기상관측['일'] = 구별_일별_기상관측.일시.map(lambda x : x.day)
구별_일별_기상관측['시'] = 구별_일별_기상관측.일시.map(lambda x : x.hour)
구별_일별_기상관측['날짜'] = 구별_일별_기상관측.일시.map(lambda x : datetime.date(x.year, x.month, x.day))

구별_일별_기상관측['구이름'] = 구별_일별_기상관측['지점'].map(lambda x : 구별_코드번호[x])
구별_일별_기상관측['구이름'] = 구별_일별_기상관측['구이름'].map(lambda x : '동작' if x=='현충원' else x)
구별_일별_기상관측['구이름'] = 구별_일별_기상관측['구이름'].map(lambda x : '종로' if x=='북악산' else x)
구별_일별_기상관측['구이름'] = 구별_일별_기상관측['구이름'].map(lambda x : x if x=='중구' else x+'구')

구별_일별_기상관측 = 구별_일별_기상관측[구별_일별_기상관측['구이름'].isin(basic_df.구이름.unique().tolist())][['년','월','일','시','구이름','기온(°C)','강수량(mm)']]
```

## 기본 DataFrame, 구별_일별_기상관측, 서울시_일별_기상관측 데이터 합치기

#### 전처리 설명 

1. 2018년 12월 31일 날씨 데이터가 없어서 검색 한 후 날씨를 맑음으로 처리
2. 강수량(1mm) 데이터로 적설량(1cm)를 계산함 ( 강수량(1mm) = 적설량(1cm) )

#### 이슈 설명

1. 기온이 없는 데이터 2263개 어떻게 처리할지
2. 강수량이 없는 데이터 4866개 어떻게 처리할지

```python
기상 = pd.merge(구별_일별_기상관측, 서울시_일별_기상관측, how='outer')
```

```python
기상.head()
```

```python
기상['기온(°C)'].isnull().sum()
```

```python
기상['강수량(mm)'].isnull().sum()
```

계산식 : 강수량(1mm) = 적설량(1cm)

```python
기상['날씨_리스트'] = 기상['날씨'].map(lambda x : x.split('/'))

기상['강설량_체크'] = 기상['날씨'].map(lambda x : '눈' in  x)
# 기상관측['강설량_체크'] = 기상관측['강설량_체크'].fillna(False)

기상['계산된적설량(cm)'] = 기상['강수량(mm)']*기상['강설량_체크']
```
