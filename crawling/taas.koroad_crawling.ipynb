```python
from bs4 import BeautifulSoup
import urllib.request as req
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import requests


import os
import glob
import shutil
```

```python
year = ['2017','2018']
place = ['강남구','강동구','강북구','강서구','관악구','광진구','구로구','금천구','노원구','도봉구','동대문구','동작구','마포구','서대문구','서초구','성동구','성북구','송파구','양천구','영등포구','용산구','은평구','종로구','중구','중랑구']
```

```python
driver = webdriver.Chrome('chromedriver')

url = 'https://taas.koroad.or.kr/web/umt/lmt/initLogin.do'

driver.get(url)

driver.find_element_by_name('loginid').send_keys('ljhljh0125')
driver.find_element_by_name('loginpwd').send_keys('dlwhdgh01!')

driver.find_element_by_xpath('//*[@id="loginbtn"]').click()

time.sleep(10)

url = 'http://taas.koroad.or.kr/gis/mcm/mcl/initMap.do?menuId=GIS_GMP_STS_RSN'

driver.get(url)

time.sleep(10)

driver.find_element_by_xpath('//*[@id="ptsRafCh1AccidentContent"]/li[2]/input').click()
driver.find_element_by_xpath('//*[@id="ptsRafCh1AccidentContent"]/li[3]/input').click()
driver.find_element_by_xpath('//*[@id="ptsRafCh1AccidentContent"]/li[4]/input').click()

time.sleep(10)

for y in year:
    
    select = Select(driver.find_element_by_id('ptsRafYearStart'))
    select.select_by_value('{}'.format(y))

    time.sleep(10)

    select = Select(driver.find_element_by_id('ptsRafYearEnd'))
    select.select_by_value('{}'.format(y))

    time.sleep(10)
    
    for p in place:
        time.sleep(10)
        
        select = Select(driver.find_element_by_id('ptsRafSigungu'))
        select.select_by_visible_text('{}'.format(p))
        
        time.sleep(10)

        driver.find_element_by_xpath('//*[@id="regionAccidentFind"]/div[2]/p/a').click()
        
        time.sleep(30)

        driver.find_element_by_xpath('//*[@id="regionAccidentFind"]/div[3]/div[2]/p/a').click()

        time.sleep(20)
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        number_data = soup.select('div.searc-total > div.total-count > span')[0].text

        driver.switch_to.window(driver.window_handles[-1])

        time.sleep(10)

        driver.find_element_by_css_selector("input[type='button']").click()

        time.sleep(60)

        filename = max(glob.iglob('/home/jongsky/Downloads/*.xls'), key=os.path.getctime)

        new_filename = filename.replace(filename, "{}.xls".format(p))

        os.rename(r'{0}'.format(filename),r'/home/jongsky/Downloads/{0}'.format(new_filename))

        shutil.move(r'/home/jongsky/Downloads/{0}'.format(new_filename), '/home/jongsky/workspace/snow_clearing/data/원본데이터/교통/구별_시간별_교통사고/{0}/{0}_{1}'.format(y,new_filename))

        driver.close()
        
        print('{0}_{1} 데이터 총 {2}개의 데이터를 가져왔습니다.'.format(y,p,number_data))

        driver.switch_to.window(driver.window_handles[-1])
        
driver.quit()
```
