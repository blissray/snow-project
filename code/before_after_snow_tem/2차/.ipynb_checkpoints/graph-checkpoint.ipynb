```python
data_list = list(set(os.listdir('서울시_온도_데이터/')))
print(data_list)
```

```python
all_df = pd.DataFrame(columns=['지점', '일시', '평균기온(°C)', '일 최심신적설(cm)', '일 최심적설(cm)'])
for i in range(len(data_list)):
    df = pd.read_csv('서울시_온도_데이터/{}'.format(data_list[i]), encoding='cp949')
    all_df = all_df.append(df)
all_df = all_df.reset_index(drop=True)
```

```python
all_df['일시'] = all_df['일시'].map(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').date())
all_df['snow_check'] = ~all_df['일 최심신적설(cm)'].isna()
all_df = all_df.sort_values(["일시"])
```

```python
all_df.head()
```

```python
# 2000년도 1월 5일은 nan값이 생김
snow_day_list = sorted(all_df[all_df['snow_check']==True].일시.unique())[1:]
snow_day_list[:5]
```

```python
len(snow_day_list)
```

```python
zero_data = [-7+i for i in range(15)]
zero_df = pd.DataFrame(zero_data, columns=['day'])

for i in snow_day_list:
    df = pd.DataFrame(all_df[(all_df['일시']<=i + timedelta(days=7))& (i - timedelta(days=7) <= all_df['일시'])]['평균기온(°C)'].reset_index(drop=True))
    df = df.rename(columns={'평균기온(°C)':i})
    zero_df = zero_df.join(df, how='outer')
    
zero_df = zero_df.set_index('day')
```

```python
zero_df
```

```python
day_list = np.concatenate([[-7+i for k in range(248)] for i in range(15)])
tem_list = np.concatenate(zero_df.values)
```

```python
data = {'day': day_list, '평균기온(°C)':tem_list}
graph_df = pd.DataFrame(data=data)
```

```python
graph_df
```

```python
ax = sns.lineplot(x='day', y='평균기온(°C)', data=graph_df, color='r', ci='sd',label='평균기온(°C)')
# ax.figure.legend()
plt.legend(fontsize=12)

ax.grid(True)

plt.title("서울시 눈 오기 전 후 기온변화(2000~2018)", fontsize=25,color='black')

ax.set_xlabel('날짜', fontsize=20, color='black')

ax.set_ylabel("평균기온(°C)",fontsize=20,color='red')

plt.rcParams["figure.figsize"] = [20, 10]
plt.rcParams['lines.linewidth'] = 1
plt.rcParams["xtick.labelsize"] = 20
plt.rcParams["ytick.labelsize"] = 20

# ax.legend(loc='lower left', bbox_to_anchor=(0.675,0.9))

fig = plt.gcf()
plt.show()
fig.savefig('결과/graph_change_std_2000-2018.png')
```

## Q. 눈이 오면 언제 다시 또 올까?
## A. 평균적으로 눈이 온다음 약 7일 3시간 뒤에 다시 눈이 온다

#### 데이터 처리시 눈이 오고 난 후에 100일 이상동안 안온경우에는 겨울이 끝난걸로 생각하고 계산에서 제거하도록 한다

```python
resnow_day_list = []
first_day = snow_day_list[0]
for i in snow_day_list[1:]:
    resnow_day_list.append(i-first_day)
    first_day = i
```

```python
resnow_day_list = [i for i in resnow_day_list if i <= timedelta(days=100) ]
```

```python
print(np.mean(resnow_day_list))
```

```python
resnow_day_list
```

```python
result = Counter(resnow_day_list)

result_sort = sorted(result.items(), key=lambda t : t[1], reverse=True)

print('눈이 오는 경우의 수 :',len(resnow_day_list),'번')
for key in result_sort:
    print(key[0].days,'일 후에 눈에 다시 오는 경우의 수 :', result[key[0]],'번')
```
